```python
import csv

f = open('2020_pan.csv', encoding='cp949')
data = csv.reader(f)

result =[]

for row in data:
  if '판교동' in row[0]:
    for i in row[3:]:
      result.append(int(i.replace(',', '')))

print(result)
print(len(result))
```

```python
import matplotlib.pyplot as plt

plt.figure(dpi=200)
plt.rc('font', family = 'NanumBarunGothic') # 나눔 바른 고딕을 기본 글꼴로 
plt.style.use('ggplot') # 격자 무늬 스타일로 지정 classic, dark_background.....
plt.title('2020년 판교 지역의 인구 구조')
plt.plot(result)
plt.show()
```
