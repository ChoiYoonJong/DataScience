```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
```

```python
df_patient = pd.read_csv('PatientInfo.csv')
```

데이터 결측지 확인

```python
df_patient.isna().sum()
```

나이 (age) -> 20s -> 23 정확하게 숫자로 변경 -> 현재년도 - 태어난년도(birth_year) <br>
1. Null(Nan) 데이터를 처리(Nan 계산 불가) -> 0으로 채우자 <br>
2. 나이 계산시, 0보다 큰 값은 (현재년도 - 태어난년도) 0보다 작으면 다시 Nan 채우기  

```python
df_patient['birth_year'] = df_patient.birth_year.fillna(0.0).astype(int)
```

```python
df_patient.isna().sum()
```

```python
#0 보다 큰 값은 그대로 사용하고, 0인 값은 Nan으로 변환 ->  나이 계산 시 2021살 방지
df_patient['birth_year'] = df_patient['birth_year'].map(lambda val : val if val > 0 else np.nan)
```

```python
df_patient['age']= 2021 - df_patient['birth_year']
df_patient.tail()
```

```python
patient = df_patient
```

```python
#격리해제자 정보 담기
released = patient[patient.state =='released']
released.info()
```

격리해제자들의 연령대 분석

```python
# 격리해제자들의 연령대 분포
plt.figure(figsize=(10,6), dpi=200)
sns.set_style('darkgrid')
plt.title("Agedistribution of the released")
sns.kdeplot(data=released['age'],shade=True)
```

03. 격리된(isolated)환자 데이터

```python
#격리된 환자의 정보 담기
isolated = patient[patient.state =='isolated']
isolated.info()
```

```python
# 격리자들의 연령대 분포
plt.figure(figsize=(10,6), dpi=200)
sns.set_style('darkgrid')
plt.title("Agedistribution of the isolated")
sns.kdeplot(data=isolated['age'],shade=True)
```

4. 사망자(deceased)데이터

```python
#사망자의 정보
dead = patient[patient.state =='deceased']
dead.info()
```

```python
# 사망자들의 연령대 분포
plt.figure(figsize=(10,6), dpi=200)
sns.set_style('darkgrid')
plt.title("Agedistribution of the decesaed")
sns.kdeplot(data=dead['age'],shade=True)
```

05. 사망자에 따른 성별(gender)데이터

```python
male_dead = dead[dead.gender=='male']
female_dead = dead[dead.gender=='female']
```

```python
male_dead.info()
```

```python
female_dead.info()
```

성별에 따른 사망자의 연령분포

```python
# 성별에 따른 사망자의 연령 분포
plt.figure(figsize=(10,6), dpi=200)
sns.set_style('darkgrid')
plt.title("Agedistribution of the decesaed the gender")
sns.kdeplot(data=male_dead['age'],label='Male',shade=True)
sns.kdeplot(data=female_dead['age'],label='Female',shade=True)
plt.legend()
```

KDE : Kernal Density Estimator 추가한 kdeplot()

```python
# 성별에 따른 사망자의 연령 분포
plt.figure(figsize=(10,6), dpi=200)
sns.set_style('darkgrid')
plt.title("Agedistribution of the decesaed the gender")
sns.distplot(a=male_dead['age'],label='Male', kde=True)
sns.distplot(a=female_dead['age'],label='Female', kde=True)
plt.legend()
```

사망자 성별 비교 -> 바차트,파이차트 -> 분포

```python
plt.figure(figsize=(10,6),dpi=200)
plt.title("Gender")
dead.gender.value_counts().plot.bar()
```

```python
plt.figure(figsize=(10,6),dpi=200)
plt.title("Gender")
dead.gender.value_counts().plot.pie(autopct = '%.f%%')
```

환자 유형별 (격리해제자,격리중인 환자,사망자) 연령 분포

```python
plt.figure(figsize=(10,6), dpi=200)
sns.set_style('darkgrid')
sns.kdeplot(data=dead['age'],label='deceased',shade=True)
sns.kdeplot(data=released['age'],label='released',shade=True)
sns.kdeplot(data=isolated['age'],label='isolated',shade=True)
plt.legend()
```
