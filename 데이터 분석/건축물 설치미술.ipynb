```python
# 1. Pandas 선언
import pandas as pd

# 2. CSV 파일 불러오기
df = pd.read_csv('/content/건축물미술작품.csv',encoding='cp949')
df.head()
```

```python
df.info()
```

지도에 표시할 건축물위치 추출

```python
df_route = df.loc[:,['분류','건축물용도','위도','경도']]
df_route.head()
```

건출물 Marker 활용하여 지도 표출

```python
# 지도에 표시하기 위해 folium 라이브러리
import folium

map = folium.Map(location=[df_route['위도'].mean(),df_route['경도'].mean()],
                 zoom_start=12)

map
```
