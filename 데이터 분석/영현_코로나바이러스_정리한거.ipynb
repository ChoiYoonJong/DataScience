```python
df['확진자'] = df['확진자'].astype(float) # 그래프에 % 표시를 위해서 float로 타입 바꿔줌

patient  = df['확진자'] # 그래프에 들어갈 숫자
country = df['국가'] # 라벨링을 위함
explode = [0.01]*9 + [0.5] + [0.01] # 9개에 explode 0.01씩, 거리 살짝식 띄워지고 10번째 한국 데이터에 0.5를 줘서 많이 띄워줌, 기타 그래프인 마지막도 0.01 추가
colors = ['dimgray', 'dimgrey', 'gray', 'darkgray', 'silver', 'lightgray', 'lightgrey', 'gainsboro', 'whitesmoke', 'red', 'gainsboro']
#색상 추가, 9개의 나라에 회색과 유사한 색상을 주어서 시선을 분산하게 했고, 한국(10번째)에 red, 기타에 회색 부여
wedgeprops={'width': 0.7}
# 파이 차트를 도넛형으로 만들기 위해 안의 원 너비 0.7, 너비 조절 가능

fontpath = '/usr/share/fonts/truetype/nanum/NanumBarunGothicBold.ttf'   # 글씨체
font = fm.FontProperties(fname=fontpath, size=9)

plt.figure(dpi=200)
plt.pie(patient, labels=country, autopct='%.2f%%', startangle=200, counterclock=False, explode=explode, colors=colors, wedgeprops=wedgeprops)
# patient (float)형 그래프여야 .%로 표시 가능
# labels=country 라벨링 이름은 나라
# autopct='%.2f%%' 파이 차트의 그래프에 % 형식으로 나타내기 위함, %%를 쓴 이유는 %하나는 포맷형식으로 인식하기 때문
# startangle=200 각도 설정, 한국을 아래 쪽에 오게 하기 위해서 원본 기준 200도 돌림
# counterclock=False 시계방향으로 설정, True로 하면 반시계방향으로 설정
# explode=explode 나라들 사이 띄워줘서 구분하기 쉽게 하기 위해서
# colors=colors 색상 설정
# wedgeprops=wedgeprops 파이 차트를 도넛형으로 만들기 위함
plt.title('전세계 코로나 현황 TOP9 + 한국\n전세계 : 180,848,208\n한국 : 153,789\t\t\t\n')
plt.axis('equal')
plt.savefig('COVID_TOP9_img.png')
# 파일 저장
plt.show()
```
