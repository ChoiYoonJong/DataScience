plt.plot(x,y) 그래프 그리기

```python
import matplotlib.pyplot as plt # 라이브러리 임포트

plt.plot([1,2,3,4],[12, 43, 25, 15]) # plt.plot(x,y)함수에 데이터 입력하기
plt.show() #그래프 보여주기
```

```python
import matplotlib.pyplot as plt # 라이브러리 임포트

plt.title('plotting') #타이틀 속성 추가
plt.plot([1,2,3,4],[12, 43, 25, 15]) # plt.plot(x,y)함수에 데이터 입력하기
plt.show() #그래프 보여주기
```

```python
import matplotlib.pyplot as plt # 라이브러리 임포트

plt.title('plotting') #타이틀 속성 추가

plt.plot([10,20,30,40], label='asc') #증가를 의미하는 asc 범례
plt.plot([40,30,20,10], label='desc') #감소를 의미하는 desc 범례

plt.legend(loc=2) #그래프에 범례 속성 추가
plt.show()
```

```python
import matplotlib.pyplot as plt # 라이브러리 임포트

plt.title('color') #타이틀 속성 추가

plt.plot([10,20,30,40], color='skyblue', label='skyblue') #하늘색 그래프
plt.plot([40,30,20,10], 'pink', label='pink') #분홍색 그래프

plt.legend() #그래프에 범례 속성 추가
plt.show()
```

```python
import matplotlib.pyplot as plt # 라이브러리 임포트

plt.title('color') #타이틀 속성 추가

plt.plot([10,20,30,40], color='c', linestyle= '--', label='dashed') # cyan dashed 그래프
plt.plot([40,30,20,10], 'pink', ls=':', label='dotted') #분홍색 dotted 그래프 linestyle -> ls 표현 가능

plt.legend() #그래프에 범례 속성 추가
plt.show()
```

```python
import matplotlib.pyplot as plt # 라이브러리 임포트

plt.title('marker') #타이틀 속성 추가

plt.plot([10,20,30,40], 'r.',  label='circle') # 빨간색 circle 그래프
plt.plot([40,30,20,10], 'g^',  label='triangle up') #초록색 triangle up 그래프 linestyle -> ls 표현 가능

plt.legend() #그래프에 범례 속성 추가
plt.show()
```

```python
import matplotlib.pyplot as plt # 라이브러리 임포트

plt.title('marker') #타이틀 속성 추가

# 'r.--' : <색상><마커모먕><선모양> 코드로 작성가능
plt.plot([10,20,30,40], 'r.-',  label='circle') # 빨간색 circle 그래프
plt.plot([40,30,20,10], 'g^--',  label='triangle up') #분홍색 dotted 그래프 linestyle -> ls 표현 가능

plt.legend() #그래프에 범례 속성 추가
plt.show()
```

데이터 시각화하기

```python
import csv 
import matplotlib.pyplot as plt # 라이브러리 임포트

f = open('/content/seoul.csv', 'r', encoding ='cp949')
data = csv.reader(f)
header = next(data) # 헤더 추출

result =[] # 결측지를 제거한 최고 기온 리스트

for row in data:
  if row[-1] != '': # 최고기온이 ''(빈문자열)이 아니면,
    result.append(float(row[-1])) 

result

plt.figure(dpi=200)
plt.plot(result,'r')
plt.show()
```
