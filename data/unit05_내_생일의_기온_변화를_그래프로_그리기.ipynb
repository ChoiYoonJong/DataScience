데이터 준비하기

```python
# 최고 기온 데이터만 리스트에 담아서 출력하기

import csv

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) #헤더

result = [] # 최고 기온 데이터를 지정할 리스트 생성

for row in data:
  if row[-1] != '' : #최고 기온 데이터 값이 존재하면 (결측치 제외)
    result.append(float(row[-1]))

print(result) # 최고기온 리스트 출력
```

데이터 개수 확인

```python
print(len(result)) # 결측치 제외 된 데이터 
```

```python
# 최고기온 값이 8월에 해당되는 값만 추출해서 그래프 시각화

import csv
import matplotlib.pyplot as plt


f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) #헤더

result = [] # 최고 기온 데이터중 8월 데이터만 지정할 리스트 생성

for row in data:
  if row[-1] != '': #최고 기온 결측치 제외
    if row[0].split('-')[1] == '08': # 날짜를 잘라서 8월 데이터만 추출
      result.append(float(row[-1])) # 결과 리스트에 담기


plt.figure(dpi=200) # 가로 10인치, 세로 2인치
plt.plot(result,'hotpink') # result 리스트의 최고 기온 데이터를 빨간색 그래프로 그리기
plt.show()
```

날짜 구간을 좁혀서 그래프 그리기

```python
import csv
import matplotlib.pyplot as plt


f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) #헤더

high = [] # 최고 기온 데이터중 11월09일 데이터만 지정할 리스트 생성
low = []

for row in data:
  if row[-1] != '' and row[-2] !='': 
    if 1988 <= int(row[0].split('-')[0]):
       if row[0].split('-')[1] == '11' and row[0].split('-')[2] == '09' :
         high.append(float(row[-1])) # 결과 리스트에 담기
         low.append(float(row[-2]))
         

# 데이터 시각화
# 차트 설정
plt.figure(dpi=200)  # 해상도
plt.rc('font', family = 'NanumBarunGothic') # 나눔 바른 고딕을 기본 글꼴로 설정
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
# plot 차트 그리기
plt.plot(high,'hotpink',label="최고 기온") #최고 기온
plt.plot(low, 'b',label="최저 기온") # 최저 기온
# 
plt.title('내 생일의 기온 변화 그래프 ')
plt.legend() #범례 표시

plt.show()
```

매년 생일 기준 최고 기온,평균 기온, 최저기온 그래프 만들기

```python
import csv
import matplotlib.pyplot as plt


f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) #헤더

high = [] # 최고 기온 데이터중 11월09일 데이터만 지정할 리스트 생성
avg = [] # 평균 기온
low = [] # 최저 기온

for row in data:
  if row[-1] != '' and row[-2] !='' and row[-3] !='': 
    if 1988 <= int(row[0].split('-')[0]):
       if row[0].split('-')[1] == '11' and row[0].split('-')[2] == '09' :
         high.append(float(row[-1])) # 결과 리스트에 담기
         low.append(float(row[-2]))
         avg.append(float(row[-3]))
         

# 데이터 시각화
# 차트 설정
plt.figure(dpi=200)  # 해상도
plt.rc('font', family = 'NanumBarunGothic') # 나눔 바른 고딕을 기본 글꼴로 설정
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
# plot 차트 그리기
plt.plot(high,'hotpink',label="최고 기온") #최고 기온
plt.plot(avg, 'g',label="평균 기온") # 평균 기온
plt.plot(low, 'skyblue',label="최저 기온") # 최저 기온

# 
plt.title('내 생일의 기온 변화 그래프 ')
plt.legend() #범례 표시

plt.show()
```
