지역명을 입력받아서 연령별 인구수 데이터 

```python
import csv

f = open('YD.csv', encoding='cp949')
data = csv.reader(f)

result =[]


#지역명을 입력받는 변수 선언
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 :')

for row in data:
  if name in row[0]:
    for i in row[3:]:
      result.append(int(i.replace(',','')))

print(result)
print(len(result))

#인구구조 시각화
import matplotlib.pyplot as plt

plt.figure(dpi=200)
plt.style.use('ggplot') # 격자 무늬 스타일로 지정 classic, dark_background.....
plt.rc('font', family='NanumBarunGothic')

plt.title(name +'지역의 인구 구조') # 입력 받은 지역 이름을 포함하여, 타이틀 속성 추가
plt.plot(result)
plt.show()
```

수평 막대그래프(barh()함수)를 활용해 우리 동네 인구 구조 시각화하기

```python
import csv

f = open('YD.csv', encoding='cp949')
data = csv.reader(f)

result =[]


#지역명을 입력받는 변수 선언
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 :')

for row in data:
  if name in row[0]:
    for i in row[3:]:
      result.append(int(i.replace(',','')))

plt.figure(dpi=200)
plt.barh(range(101),result)
plt.title(name +'지역의 인구 구조') # 입력 받은 지역 이름을 포함하여, 타이틀 속성 추가
plt.show()
```

성별 데이터 저장하가(Type1)

```python
import csv

f = open('gender.csv', encoding='cp949')
data = csv.reader(f)

m = []
f = []

for row in data:
  if '신도림' in row[0]:
    for i in range(0,101) :
      m.append(int(row[i+3]))
      f.append(int(row[-(i+1)])) # -1 값부터 담음 -> 100세 이상 부터 값이 담김

f.reverse() # 0세 부터 인구수 나올 수 있도록 데이터 뒤집기

print(m)
print(f)
```

성별 데이터 순서대로 저장하가(Type2)

```python
import csv

f = open('gender.csv', encoding='cp949')
data = csv.reader(f)

m = []
f = []

for row in data:
  if '신도림' in row[0]:
    for i in row[3:104] : # 남성 데이터 : 3번 ~ 103번 데이터 슬라이싱 - 정수 데이터로 변환 - m 리스트에 저장
      m.append(int(i))
    for i in row[106:]:   # 여성 데이터 : 3번 ~ 103번 데이터 슬라이싱 - 정수 데이터로 변환 - f 리스트에 저장
      f.append(int(i))

print(m)
print(f)
```

데이터 시각화하기

```python

import matplotlib.pyplot as plt

plt.figure(dpi=200)
plt.rc('font', family = 'NanumBarunGothic')
plt.barh(range(101), m)
plt.barh(range(101), f)
plt.show()
```

```python
import csv

f = open('gender.csv', encoding='cp949')
data = csv.reader(f)

m = []
f = []

for row in data:
  if '신도림' in row[0]:
    for i in row[3:104] : # 남성 데이터 : 3번 ~ 103번 데이터 슬라이싱 - 정수 데이터로 변환 - m 리스트에 저장
      m.append(-int(i))
    for i in row[106:]:   # 여성 데이터 : 3번 ~ 103번 데이터 슬라이싱 - 정수 데이터로 변환 - f 리스트에 저장
      f.append(int(i))

print(m)
print(f)
```

```python
import matplotlib.pyplot as plt

plt.figure(dpi=200)
plt.rc('font', family = 'NanumBarunGothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('신도림 지역의 남녀 성별 인구 분포')
plt.barh(range(101), m, label = '남성') # Make a horizontal ba plot,
plt.barh(range(101), f, label ='여성') # Make a horizontal ba plot,
plt.legend()
plt.show()
```
