꺽은선 그래프로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv', encoding='cp949')
data = csv.reader(f)

m = []
f = []

name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 :')

for row in data:
  if name in row[0]:
    for i in row[3:104] :
      m.append(int(i.replace(',','')))
    for i in row[106:]:
      f.append(int(i.replace(',',''))) # -1 값부터 담음 -> 100세 이상 부터 값이 담김
    break



plt.style.use('ggplot')
plt.figure(dpi=200)
plt.rc('font', family = 'NanumBarunGothic')

#꺽은선 그래프로 남여 인구수 데이터 표현하기 
plt.plot(m, label='남성인구수')
plt.plot(f, label='여성인구수')
plt.legend()
plt.show()
```

연령별로 남성 인구수에서 여성 인구수를 뺀 차이의 막대그래프 그리기

```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv', encoding='cp949')
data = csv.reader(f)

result= [] # 남성 인구 - 여성 인구 값 담을 리스트 변수 선언 


name = input('찾고 싶은 지역의 아룸울 얼려주세요 :')

for row in data:
  if name in row[0]:
    for i in range(3,104):
      result.append(int(row[i].replace(',','')) - int(row[i+103].replace(',','')))
    break


plt.style.use('ggplot')
plt.figure(dpi=200)
plt.rc('font', family = 'NanumBarunGothic')
plt.rcParams['axes.unicode_minus'] = False

plt.title(name + '지역의 남녀 성별 인구차이')
plt.bar(range(101),result)
plt.show()
```

원의 크기는 인구수의 제곱근 추가

```python
import csv
import matplotlib.pyplot as plt
import math

f = open('gender.csv', encoding='cp949')
data = csv.reader(f)

m = []
f = []
size =[] # 인구수 사이즈 리스트 변수 선언

name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 :')

for row in data:
  if name in row[0]:
    for i in row[3:104] :
      m.append(int(i.replace(',','')))
      size.append(math.sqrt(int(i.replace(',',''))))
    
    for i in row[106:]:
      f.append(int(i.replace(',','')))
    break


#산점도 그래프로 남여 인구수 데이터 표현하기 
plt.figure(dpi=200) 
plt.scatter(m,f,s=size,c=range(101), cmap='jet',alpha=0.5) #x 축은 남자, y축은 여자
plt.colorbar()
plt.plot(range(max(m)),range(max(m)), 'g')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
import math

f = open('gender.csv', encoding='cp949')
data = csv.reader(f)

m = []
f = []
size =[] # 인구수 사이즈 리스트 변수 선언

name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 :')

for row in data:
  if name in row[0]:
    for i in row[3:104] :
      m.append(int(i.replace(',','')))
      size.append(math.sqrt(int(i.replace(',',''))))
    
    for i in row[106:]:
      f.append(int(i.replace(',','')))
    break

#인구수와 sqrt(인구수) 차이 확인
print(m)
print(size)


#산점도 그래프로 남여 인구수 데이터 표현하기 
plt.figure(dpi=200) 
plt.scatter(m,f,size,c=range(101), cmap='jet',alpha=0.5) #x 축은 남자, y축은 여자
plt.colorbar()
plt.plot(range(max(m)),range(max(m)), 'g') #추세선(기준 ) 추가
plt.show()
```
