```python
# 1. Pandas 선언
import pandas as pd

# 2. CSV 파일 불러오기
df_seoul = pd.read_csv('/content/건축물미술작품_서울.csv',encoding='cp949')
df_seoul.head()
```

```python
df_seoul.info()
```

지도에 표시할 미술장식 정보 추출

```python
df_artworks = df_seoul.loc[:,['지역','건축물용도','경도','위도']]
df_artworks.head()
```

```python
# 지도에 표시하기 위해 folium 라이브러리
import folium

map = folium.Map(location=[df_artworks['위도'].mean(),df_artworks['경도'].mean()],
                 zoom_start=11)

map
```

```python
# 지도에 표시하기 위해 folium 라이브러리
import folium

map = folium.Map(location=[df_artworks['위도'].mean(),df_artworks['경도'].mean()],
                 zoom_start=11)

for n in df_artworks.index:

  if df_artworks.loc[n, '건축물용도'] == '공동주택':
    folium.Marker([df_artworks.loc[n, '위도'],df_artworks.loc[n, '경도']], # 위도 , 경도
                popup=df_artworks.loc[n, '지역']+''+ df_artworks.loc[n, '건축물용도'],
                icon=folium.Icon(color='darkred', icon='star', prefix='fa')).add_to(map)
  
  elif df_artworks.loc[n, '건축물용도'] == '업무시설':
    folium.Marker([df_artworks.loc[n, '위도'],df_artworks.loc[n, '경도']], # 위도 , 경도
                popup=df_artworks.loc[n, '지역']+''+ df_artworks.loc[n, '건축물용도'],
                icon=folium.Icon(color='blue', icon='star', prefix='fa')).add_to(map)
  
  elif df_artworks.loc[n, '건축물용도'] == '숙박시설':
    folium.Marker([df_artworks.loc[n, '위도'],df_artworks.loc[n, '경도']], # 위도 , 경도
                popup=df_artworks.loc[n, '지역']+''+ df_artworks.loc[n, '건축물용도'],
                icon=folium.Icon(color='red', icon='star', prefix='fa')).add_to(map)

  elif df_artworks.loc[n, '건축물용도'] == '판매시설':
    folium.Marker([df_artworks.loc[n, '위도'],df_artworks.loc[n, '경도']], # 위도 , 경도
                popup=df_artworks.loc[n, '지역']+''+ df_artworks.loc[n, '건축물용도'],
                icon=folium.Icon(color='green', icon='star', prefix='fa')).add_to(map)
  
  elif df_artworks.loc[n, '건축물용도'] == '의료시설':
    folium.Marker([df_artworks.loc[n, '위도'],df_artworks.loc[n, '경도']], # 위도 , 경도
                popup=df_artworks.loc[n, '지역']+''+ df_artworks.loc[n, '건축물용도'],
                icon=folium.Icon(color='black', icon='star', prefix='fa')).add_to(map)

  elif df_artworks.loc[n, '건축물용도'] == '근린생활시설':
    folium.Marker([df_artworks.loc[n, '위도'],df_artworks.loc[n, '경도']], # 위도 , 경도
                popup=df_artworks.loc[n, '지역']+''+ df_artworks.loc[n, '건축물용도'],
                icon=folium.Icon(color='black', icon='star', prefix='fa')).add_to(map)

  else:
    folium.Marker([df_artworks.loc[n, '위도'],df_artworks.loc[n, '경도']], # 위도 , 경도
                popup=df_artworks.loc[n, '지역']+''+ df_artworks.loc[n, '건축물용도'],
                icon=folium.Icon(color='pink', icon='circle', prefix='fa')).add_to(map)
map
```
