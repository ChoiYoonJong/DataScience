```python
import csv

f = open('population.csv', encoding='cp949')
data = csv.reader(f)

m = []
f = []

name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 :')

for row in data:
  if name in row[0]:
    for i in range(0,101) :
      m.append(-int(row[i+3]))
      f.append(int(row[-(i+1)])) # -1 값부터 담음 -> 100세 이상 부터 값이 담김

f.reverse() # 0세 부터 인구수 나올 수 있도록 데이터 뒤집기

print(m)
print(f)

import matplotlib.pyplot as plt

plt.figure(dpi=200)
plt.rc('font', family = 'NanumBarunGothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('17년09월노량진의남녀비율')
plt.barh(range(101), m,color='b', label = '남성') # Make a horizontal ba plot,
plt.barh(range(101), f, color='r', label ='여성') # Make a horizontal ba plot,
plt.legend()
plt.show()


```

```python
import csv

f = open('population.csv', encoding='cp949')
data = csv.reader(f)

m = []
f = []

name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 :')

for row in data:
  if name in row[0]:
    for i in range(0,101) :
      m.append(int(row[i+3]))
      f.append(int(row[-(i+1)])) # -1 값부터 담음 -> 100세 이상 부터 값이 담김

f.reverse() # 0세 부터 인구수 나올 수 있도록 데이터 뒤집기

plt.figure(dpi=200)
plt.rc('font', family = 'NanumBarunGothic')
plt.rcParams['axes.unicode_minus'] = False
plt.bar(range(0,101),m,color='b',label='남성')
plt.bar(range(0,101),f, color='r',label='여성')
plt.title('17년09월노량진의남녀비율') # 입력 받은 지역 이름을 포함하여, 타이틀 속성 추가
plt.legend()
plt.show()


```
