```python
import plotly.express as px

import pandas as pd
```

데이터 준비하기

```python
df = pd.read_csv('/content/traffic_accident.csv', encoding='euc-kr')
df.head()
```

```python
df.info()
```

교통사고 시간 전처리

```python
#Substr 문자열 자르기 위해  데이터타입 변경
df = df.astype({'발생년월일시' : 'string'})
```

```python
df.info()
```

```python
#2019010100 -> 00 시간으로 분리
df['발생시간'] = df['발생년월일시'].str[8:]
```

```python
df.info()
```

```python
#발생시간 숫자 데이터타입으로 변경
df = df.astype({'발생시간' : 'int64'})
```

```python
df.info()
```

```python
#2019010100 -> 20190101 추출 -> 2019-01-01 (날짜형태 YYYY-MM-DD 형태로 변경)
df['발생년월일시'] = pd.to_datetime(df['발생년월일시'].str[:8],
                              format='%Y-%m-%d',errors='raise' )
```

```python
df.info()
```

```python
df.head()
```

지도 활용하여 교통사고 현황 분석

지도에 필요한 정보만 담기 -> 충북의 교통사고 현황을 지도에 표출

```python
df_map1 = df[df.발생지시도=='충북']
df_map1.info()
```

```python
df_map1 = df_map1.astype({'사망자수': 'float64'})
df_map1 = df_map1.astype({'경상자수': 'float64'})
df_map1 = df_map1.astype({'중상자수': 'float64'})
df_map1 = df_map1.astype({'부상자수': 'float64'})
```

```python
df_map1.info()
```

사망자수, 경상자 수에 따른 CircleMarker 지도에 표출하기

```python
for m in df_map1.index:
  cnt = df_map1['사망자수'][m] + df_map1['경상자수'][m] #사망자수 + 경상자수 -> CircleMaker 사이즈
  folium.CircleMarker([df_map1['위도'][m], df_map1['경도'][m]],
                      radius=cnt*10, popup=df_map1['사고유형'][m], color='#3186cc', fill_color='#3186cc').add_to(map)
map.save('충북교통사고현황.html')
```
