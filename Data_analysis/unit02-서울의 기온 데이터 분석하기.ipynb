헤더 저장하기

```python
import csv

f = open('/content/seoul.csv', 'r', encoding ='cp949')
data = csv.reader(f)
header = next(data)
print(header) #첫 번째 행을 읽은 후 다음 행으로 이동 시키는 함수
f.close()
```

```python
import csv

f = open('/content/seoul.csv', 'r', encoding ='cp949')
data = csv.reader(f)

header = next(data)
print(header) #첫 번째 행을 읽은 후 다음 행으로 이동 시키는 함수

print(next(data)) # 두번째 행 읽은 후 다음 행으로 이동
f.close()
```

```python
import csv

f = open('/content/seoul.csv', 'r', encoding ='cp949')

data = csv.reader(f)
header = next(data)
print(header) # 헤더 추출

# 데이터를 한줄씩 출력
for row in data:
  print(row)

f.close()
```
