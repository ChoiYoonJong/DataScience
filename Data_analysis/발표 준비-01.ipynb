```python
# 1. Pandas 선언
import pandas as pd

# 2. CSV 파일 불러오기
df_wholecountry = pd.read_csv('/content/건축물미술작품_전국.csv',encoding='cp949')
df_wholecountry.head()
```

```python
df_wholecountry.info()
```

라이브러리 임포트

```python
import plotly.express as px

import numpy as np
import pandas as pd
```

```python
fig_wholecountry = px.bar(df_wholecountry, x='분류', y='지역',color='건축물용도',
             title='전국 미술장식 분류 현황')
fig_wholecountry.show()
```

```python
fig_wholecountry = px.bar(df_wholecountry, x = '분류', y = '건축물용도', orientation = 'h',color = '지역',
             title = '전국 미술장식 분류 현황')
fig_wholecountry.show()
```

```python
fig_wholecountry.write_image('전국 미술장식 분류 현황.png')
```

```python
fig_wholecountry =px.scatter(df_wholecountry, x='분류', y='설치일자', color='지역')
fig_wholecountry.show()
```
